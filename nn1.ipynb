```python
# Simulating learning to throw paper into a trash can

target_distance = 3.0
throw_strength = 2.0
learning_rate = 0.2
input_strength = 1.0
num_throws = 30

print("Throw | Strength | Distance | Error")
print("------------------------------------")

for i in range(1, num_throws + 1):
    distance = throw_strength * input_strength
    error = target_distance - distance
    throw_strength += learning_rate * error

    print(f"{i:5d} | {throw_strength:8.3f} | {distance:8.3f} | {error:6.3f}")
```

```python
# pouring water into a cup

target_volume = 250.0   # ml
hand_tilt = 10.0        # initial tilt angle
learning_rate = 0.3
input_tilt = 1.0
num_pours = 30
tolerance = 0.1

print("Pour | Tilt | Volume | Error")
print("-------------------------------")

for i in range(1, num_pours + 1):
    # Simple model: more tilt = more water
    poured_volume = hand_tilt * input_tilt * 10

    error = target_volume - poured_volume

    # adjust tilt based on error
    hand_tilt += learning_rate * error / 10

    print(f"{i:4d} | {hand_tilt:5.2f} | {poured_volume:7.2f} | {error:6.2f}")

    # stop early if we're close enough
    if abs(error) <= tolerance:
        print("\nTarget volume reached within tolerance!")
        break
```

```python
# setting alarm volume

target_volume = 80.0
current_volume = 60.0
learning_rate = 0.3
input_knob = 1.0
num_adjustments = 30
tolerance = 0.01

for step in range(num_adjustments):
    error = target_volume - current_volume
    adjustment = learning_rate * error * input_knob
    current_volume += adjustment

    print(f"Step {step+1}: volume = {current_volume:.2f}")

    if abs(error) < tolerance:
        print("Target volume reached within tolerance.")
        break

```
